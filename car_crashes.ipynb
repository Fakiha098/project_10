# car_crashes dataset

 ## import libarires

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load dataset

```python
# Load the car_crashes dataset
crashes = sns.load_dataset("car_crashes")
crashes
```

```python
# Display the first few rows of the dataset
print(crashes.head())
```

```python
# Get basic information about the dataset
print(crashes.info())
```

```python
# Get descriptive statistics for numerical columns
print(crashes.describe())
```

```python
# Check for missing values
print(crashes.isnull().sum())
```

### viz the total Accidents

```python
# Plot the distribution of total accidents
plt.figure(figsize=(8, 5))
sns.histplot(crashes['total'], bins=20, kde=True)
plt.title("Distribution of Total Car Accidents")
plt.xlabel("Total Accidents")
plt.ylabel("Frequency")
plt.show()
```

## Heatmap

```python
# Plot the correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(crashes.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Car Crashes Dataset")
plt.show()
```

```python

# Scatter plot for speeding vs total accidents
plt.figure(figsize=(8, 5))
sns.scatterplot(data=crashes, x="speeding", y="total", hue="alcohol", palette="coolwarm")
plt.title("Speeding vs Total Accidents (Colored by Alcohol Involvement)")
plt.xlabel("Speeding (% of accidents)")
plt.ylabel("Total Accidents")
plt.show()
```

```python
# Bar plot for total accidents by state
plt.figure(figsize=(10, 6))
sns.barplot(data=crashes, x="abbrev", y="total")
plt.title("Total Accidents by State")
plt.xlabel("State")
plt.ylabel("Total Accidents")
plt.xticks(rotation=90)
plt.show()
```

```python
# Pairplot for car crashes dataset
sns.pairplot(crashes)
plt.show()

```

```python
# Box plot for alcohol involvement by state
plt.figure(figsize=(12, 6))
sns.boxplot(data=crashes, x="abbrev", y="alcohol")
plt.title("Alcohol Involvement in Car Accidents by State")
plt.xlabel("State")
plt.ylabel("Alcohol (% of accidents)")
plt.xticks(rotation=90)
plt.show()
```

 ## Conclusions

```python


print("Conclusions:")
print("- There is a positive correlation between speeding and total accidents.")
print("- Alcohol involvement varies significantly by state.")
print("- States with higher percentages of speeding and alcohol involvement tend to have more accidents.")
```
